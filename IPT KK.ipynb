## SLICING A DATA

### DROP A COLUMN "Cost of Cultivation (`/Hectare) A2+FL"  

```python
agric=agri.drop(['Cost of Cultivation (`/Hectare) A2+FL'],axis=1)
```

```python
agric
```

```python
agricr=agric.rename(columns={'Cost of Cultivation (`/Hectare) C2':'Cost of Cultivation','Cost of Production (`/Quintal) C2':'Cost of Production'})
```

```python
agricr
```

```python
x=agricr.iloc[:,2].values.reshape(-1,1)
y=agricr.iloc[:,-2].values.reshape(-1,1)
```

## SCATTER PLOT

```python
plt.scatter(x,y)
plt.xlabel('Cost of Cultivation')
plt.ylabel('Cost of Production')
plt.show()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

## Fitting Multiple Linear Regression to the Training set

```python
from sklearn.linear_model import LinearRegression
```

```python
ln=LinearRegression()

ln
```

```python
ln.fit(x_train,y_train)
```

## Predicting the Test set results

```python
y_predict=ln.predict(x_test)
```

```python
check=pd.DataFrame(x_test,columns=[' Cultivation'])
```

```python
check['Actual Cost of Production ']=y_test
```

```python
check['  Prediction']=y_predict
```

```python
check
```

```python
plt.scatter(x,y,color='purple')
plt.plot(x_test,y_predict,color='red')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mean_squared_error(y_test,y_predict)
```

```python
mean_absolute_error(y_test,y_predict)
```

```python
r2_score(y_test,y_predict)
```
